# Basic Function

### Import libraries

```python
import numpy as np
import pandas as pd

from pandas import Series, DataFrame
```

### Creat a seties object as 'so' and a dataframe object as 'df' for examples

```python
so = Series(np.arange(8),index=['row 1','row 2','row 3','row 4','row 5','row 6','row 7','row 8'])
```

```python
df = DataFrame(np.random.rand(36).reshape((6,6)),
                  index=['row 1','row 2','row 3','row 4','row 5','row 6'],
                  columns=['column 1','column 2','column 3','column 4','column 5','column 6'])
```

### Select specific rows and/or columns 

```python
# use double square brackets if you have a list inside a list.
print(so['row 7'])
print(so[[0,7]])
print(so['row 3':'row 7'])
print(so[so > 6])
```

```python
# select and return a slice of several values from a data set
df.loc[['row 2','row 5'],['column 5','column 2']]
```

### Figuring out missing data

```python
so.isnull()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
# drop NA according to rows (drop row if any NA appear in that row)
df.dropna()
```

```python
# drop NA according to columns (drop column if any NA appear in that column)
df.dropna(axis = 1)
```

### Deal with duplicate

```python
# rule: return True if each column values in that row as same as any row before.
df.duplicated()
```

```python
# find total duplicates
df.duplicated().sum()
```

```python
# drop duplicates
df.drop_duplicates()
```

```python
# only drop duplicates between select column.
df.drop_duplicates(['column 3'])
```

### Concatenating dataframe

```python
# concatenate by rows
pd.concat([df,df])
```

```python
# concatenate by columns
pd.concat([df,df],axis=1)
```

### Drop data

```python
df.drop(['row 1','row 2']) #drop zero and second rows from tha dataframe.
```

```python
df.drop(['column 1','column 2'],axis=1) #drop zero and second columns from tha dataframe.
```

### Add data

```python
# before add any series, make sure name it first.
so.name = 'new_col'
DataFrame.join(df,so)
```

```python
df.append(df,ignore_index=True)
# same result as concat
# if ignore_index is False then index will remain same
```

### Sort data

```python
df.sort_values(by=('column 1'),ascending=False) 
# sort ascending vs. descending, default True, false is in decreasing order
```

### Group by

```python
df.groupby(df['column 1']).mean()
```
