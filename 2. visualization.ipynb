# Data Visualization

### Import libraries 

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
from pandas.plotting import scatter_matrix

import matplotlib.pyplot as plt
from matplotlib import rcParams

import seaborn as sb
sb.set_style('whitegrid')
# let chart has a white grid.

%matplotlib inline
rcParams['figure.figsize'] = (10,8) 
# this will set the figure size to 10 x 8.
```

### Creat a dataframe object as 'df' for examples

```python
df = DataFrame(np.random.rand(36).reshape((6,6)),
                  columns=['column 1','column 2','column 3','column 4','column 5','column 6'])
df
```

### Line chart

```python
plt.subplot(121)
plt.plot(df['column 1'])

plt.subplot(122)
c1 = df['column 1']
c1.plot()
# different way for same plot
```

```python
df.plot()
```

### Bar chart

```python
index=['row 1','row 2','row 3','row 4','row 5','row 6']
plt.bar(x=index,height=df['column 1'])
```

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'barh')
# change bar to barh makes the bar horizontally.
```

### Pie chart

```python
plt.pie(df['column 1'])
plt.show()
```

### Histograms

```python
plt.subplot(121)
df['column 4'].plot(kind='hist')
plt.subplot(122)
plt.hist(df['column 4'])
plt.show()
```

### Scatter plot

```python
df.plot(kind='scatter',x='column 1',y='column 3',c=['lavender'],s=300)
```

### Box plot

```python
so = Series([1,1,0,1,0,0])
so.name = 'new_col'
df1 = DataFrame.join(df,so)
df1.boxplot(column='column 1',by='new_col')
```

### Add details

#### Color

```python
color = ['lightpink','#77e0cd','#b3ccf5','#ffc0b3','seagreen','#ff94fd']
df.plot(color = color)
```

```python
index=['row 1','row 2','row 3','row 4','row 5','row 6']
plt.bar(x=index,height=df['column 1'], color = color)
```

```python
df.plot(kind = 'bar',color=color)
```

```python
plt.pie(df['column 1'],colors=color)
plt.show()
```

#### Line styles

```python
plt.plot(df['column 1'],drawstyle='steps',lw=5) ## use drawstyle if ds='steps'shows a AttributeError.
plt.plot(df['column 2'],ls='--',lw=3)
```

```python
plt.plot(df['column 3'],marker='1',mew=50,color='lightpink')
plt.plot(df['column 4'],marker='+',mew=20,color='turquoise')
plt.plot(df['column 5'],marker='8',mew=10,color='orange')
```

#### Labels

```python
index=['row 1','row 2','row 3','row 4','row 5','row 6']
plt.pie(df['column 1'],colors=color)
plt.legend(index,loc='best')
plt.show()
```

```python
# make a canvas 
fig, ax = plt.subplots(figsize=(10,8))
# make plots
ax.plot(df['column 2'],color='lightpink')
ax.plot(df['column 5'],color='turquoise')
# add details
ax.set_xticks(range(6))
ax.set_xticklabels(index, rotation = 60,fontsize = 'medium')
ax.set_title('This is an example plot',fontsize=20)
ax.set_xlabel('rows', fontsize=20,fontfamily = 'sans-serif',fontstyle='italic')
ax.set_ylabel('columns', fontsize='x-large',fontstyle='oblique')
ax.legend(loc='best')
```

### Use seaborn

```python
sb.distplot(df['column 1'])
```

```python
sb.regplot(x='column 1',y='column 3',data = df,scatter=True)
```

```python
sb.pairplot(df[['column 1','column 2','column 3']])
```

```python
so = Series([1,1,0,1,0,0])
so.name = 'new_col'
df1 = DataFrame.join(df,so)

sb.boxplot(x='new_col',y='column 1',data=df1,palette='hls')
```
